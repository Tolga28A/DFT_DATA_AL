```python
import matplotlib.pyplot as plt

dft, ml = [], []

for line in open('force_correlation.dat', 'r'):
  values = [float(s) for s in line.split()]
  dft.append(values[0])
  ml.append(values[1])

plt.scatter(dft, ml)
plt.title('Force correlation between DFT and MLIP')
plt.xlabel('DFT force (eV/Angstrom)')
plt.ylabel('Machine Learning Force (eV/Angstrom)')
plt.axis('equal')
plt.xlim(-0.75,1)
plt.ylim(-0.75,1)
plt.show()
```
